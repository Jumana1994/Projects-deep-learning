Breast cancer Dataset
dataset based on breast cancer  detection is how to classify tumors into malignant (cancerous) or benign(non cancerous). 
link:https://www.kaggle.com/datasets/yasserh/breast-cancer-dataset


```python
import sklearn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB,BernoulliNB,MultinomialNB
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
data=pd.read_csv("breast-cancer.csv")
data
```

DATA PREPROCESSING
In data preprocessing we are substititing 'missing' values with na values and then filling it with bfill or ffill and also taking mean value

```python
data.isnull().any()
```

Label encoder
important preprocessing step
SKlearn efficient tool for encoding labels into numeric values
Prediction output is 'diagnosis' and converted to numeric values for prediction

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
```

```python
diagnosis=label.fit_transform(data["diagnosis"])
```

```python
data["diagnosis"]=diagnosis
```

```python
data
```

output to be predicted is 'diagnosis' whether cancer is Beneign or malignant,so x is given values other than diagnosis and y is diagnosis

```python
x=data.drop(["diagnosis"],axis=1)
y=data["diagnosis"]
```

```python
Model creation
model is prepared and cleaned for machine learning prediction
Model selection
we are checking score with available classification models andmodel with high score is taken for further process
```

```python
result=[]
models={'logistic':LogisticRegression(),'decision':DecisionTreeClassifier(),'random':RandomForestClassifier()}
for i in models:
    model=models[i]
    scores=cross_val_score(model,x,y,cv=5)
    a=(i,scores.mean())
    result.append(a)
```

```python
result
```

```python
models1={'gaussian':GaussianNB(),'bernoulli':BernoulliNB(),'multinomial':MultinomialNB()}
for i in models1:
    model=models1[i]
    scores=cross_val_score(model,x,y,cv=5)
    b=(i,scores.mean())
    result.append(b)
```

```python
result
```

Random forest classifier is getting high score then we can do train test split to estimate the performance of ML algorithms.

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=.8,random_state=1)
model=RandomForestClassifier()
model.fit(xtrain,ytrain)
model.score(xtest,ytest)
```

```python
we got 95 % score
Confusion matrix
it is a table used to define the performance of a classification algorithm
```

```python
y_act=ytest
y_predict=model.predict(xtest)
from sklearn.metrics import confusion_matrix
z=confusion_matrix(y_act,y_predict)
from mlxtend.plotting import plot_confusion_matrix
plot_confusion_matrix(z)
```
